```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline       
```

Se sube la data APP


```
df = pd.read_excel('APP.xlsx')
df
```

**Resumen de la data**

```
df.info()
```

**Gráfica Rugplot** variable recomendación P.1

Gráfica displot

```
sns.distplot(df['P.1'], kde=True)  
```

**Gráficas Recomendación Vs Satisfacción**
Jointplot relación entre dos columnas numéricas

```
hexplot = sns.jointplot(x=df["P.1"], y=df["P.3"], kind="hex")
plt.subplots_adjust(left=0.2, right=0.8, top=0.8, bottom=0.2) 
cbar_ax = hexplot.fig.add_axes([.85, .25, .05, .4])  # x, y, width, height
plt.colorbar(cax=cbar_ax)
plt.show()
```

```
sns.jointplot(x='P.1',y='P.3',data=df)
```

**Gráfico Barplot**

```
calif=df['P.1'].apply(lambda x: ((x-1)/9*100))
sns.barplot(x='SEGMENTO',y=calif,data=df) 
plt.ylabel('Recomendación', size=13)
```

```
plt.figure(figsize=(12,3))
sns.countplot(x='P.38', palette='dark', data=df)
plt.xlabel('Estrato', size=12) 
```

```

plt.figure(figsize=(18,3))
ax = sns.boxplot(x="LISTA DE ATRIBUTOS", y="P.1", data=df, palette='Set1')
ax.set_title("BOXPLOT")
ax.set_xlabel("Atributos")
ax.set_ylabel("Recomendación")
```

```
sns.boxplot(x=' Producto', y='P.1', data=df, hue='SEGMENTO', palette='tab10')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize= 10)
```

```
sns.violinplot(x=' Producto', y='P.1', data=df, hue='SEGMENTO')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize= 10)
```

**stripplot**
Diagramas de dispersión

```
plt.figure(figsize=(10,3))
sns.stripplot(x="P.38", y="P.1", data=df)
```

```
sns.stripplot(x=" Producto", y="P.1", data=df, jitter=False)
```

```
sns.stripplot(x=' Producto', y='P.1', data=df, hue='SEGMENTO', palette='Set1', dodge=True)
```
